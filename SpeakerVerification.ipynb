```python
import wespeaker

print("Đang nạp mô hình WeSpeaker (Python 3.9)...")
# Tải mô hình nhận diện giọng nói cực mạnh
model = wespeaker.load_model('english')
model.set_device('cpu')
```

```python
FILE_DANG_KY = "vinh_voice_smartphone.wav" 
FILE_KIEM_TRA = "test_mic.wav" 

print("Đang so sánh chữ ký giọng nói...")
try:
    similarity_score = model.compute_similarity(FILE_DANG_KY, FILE_KIEM_TRA)
    
    print("\n" + "="*40)
    print(f" ĐIỂM TƯƠNG ĐỒNG (COSINE): {similarity_score:.4f}")
    
    # Ngưỡng an toàn thường là 0.5
    if similarity_score >= 0.5:
        print(" ✅ KẾT QUẢ: ACCEPTED (Giọng nói khớp. Kích hoạt relay mở cửa!)")
    else:
        print(" ❌ KẾT QUẢ: REJECTED (Phát hiện xâm nhập, từ chối mở cửa!)")
    print("="*40)
        
except Exception as e:
    print(f"Lỗi: {e}")
```
